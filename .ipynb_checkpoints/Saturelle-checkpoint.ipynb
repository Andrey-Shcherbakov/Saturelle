### S1. 
Écrire un programme qui prend en entrée un couple (p, q) avec p ≤ q et qui renvoie un tableau
numpy de taille p+q qui code un chemin de (0, 0) à (p, q) uniforme parmi tous les chemins possibles. On
pourra d’abord coder une fonction qui code (par convention) Nord=1 et Est=0 dans le tableau numpy.
Indication: on pourra utiliser la méthode np.random.permutation pour générer une permutation
aléatoire uniforme. 

```python
def chemin(p, q):
    if p > q:
        raise Exception("Sorry, we need p <= q")
    
    chemin = np.zeros(p+q, dtype=int)

    for i in range(q):
        chemin[i] = 1
    
    chemin = np.random.permutation(chemin)
        
    return chemin
```

### S2.
Vérifier la validité du code précédent en écrivant un code Python qui affiche 5 chemins aléatoire

dernier pour (p = 1000, q = 1000). Que constatez-vous, pouvez-vous l’expliquer (informellement) ?

```python
p = (int)(input())
q = (int)(input())

for i in range (5):
    print(chemin(p, q))
```

```python
def graphique (chemin):
    
    x = np.zeros(chemin.size+1)
    y = np.zeros(chemin.size+1)
    for i in range(chemin.size):
        if(chemin[i]):
            x[i+1] = x[i]
            y[i+1] = y[i]+1
        else:
            x[i+1] = x[i]+1
            y[i+1] = y[i]
            
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot()
    ax.step(x, y, markersize=6, color = 'g', linewidth = 2)
    #ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    #ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    plt.yticks(fontsize=12)
    plt.xlabel(r'p', fontsize=20)
    plt.ylabel(r'q', fontsize=20)
    plt.grid(which='both')
    #plt.title(r'(%i, %i)' %(p, q), fontsize=20)
```

### S3.
Implémenter une fonction qui prend en entrée un couple (p, q) avec p ≤ q et qui renvoie un
chemin de (0,0) à (p, q) qui ne traverse pas la première bissectrice en utilisant la méthode du rejet
(qui consiste à générer des chemins avec le code S1 et à conserver le premier qui ne traverse pas la
première bissectrice). Justifier rapidement que cette algorithme de rejet renvoie bien un chemin qui suit
la mesure uniforme sur les chemins qui ne traversent pas la première bissectrice.

```python
def verif(chemin):
    sum = chemin[0]
    for i in range(chemin.size-1):
        if(i > 0 and sum <= (i-i%2)/2):
            #print(i)
            return False
        sum = sum + chemin[i+1]
    return True

def rejet(p, q):
    if p > q:
        raise Exception("Sorry, we need p <= q")
    R = 0
    ch = chemin(p, q)
    while not verif(ch):
        ch = chemin(p, q)
        R = R+1
    
    print(R)
    return ch
        
```

### S5.
Notons R le nombre de rejets réalisés lors de la génération d’un chemin qui ne traverse pas la
première bissectrice selon l’algorithme implémenté en S3. D’après la question T4, quelle est la loi de
R ? Vérifier numériquement ce résultat en générant un grand nombre de tels chemin et en affichant
l’histogramme de la loi empirique de R que l’on superposera avec la loi théorique. On affichera les
résultats pour (p = 10, q = 10) puis pour (p = 20, q = 40).

#### Solution
On a de point **T4** que la formule $P(A) = \frac{q+1-p}{q+1}$ de ça on peut facilement déduire que $P(R) = (1-P(A))^{R-1}*P(A)$ : le R-ième iteration d'algorithme nous donne le **premiére** fois, quand il ne traverse pas la bissecrtice - c'est à dire que il ne traverse pas le premiére R-1 fois et traverse le R-iéme 

```python
def R_loi(R, p, q):
    proba = (q+1-p)/(q+1)
    return (1 - proba)**(R-1) * proba


def R_graphique(p, q):
    if p > q:
        raise Exception("Sorry, we need p <= q")
    
    R_data = np.zeros(1000)
    Rmax = 0
    
    for i in range(1000):
        R = 0
        ch = chemin(p, q)
        while not verif(ch):
            ch = chemin(p, q)
            R = R+1
        if Rmax < R:
            Rmax = R
        R_data[i] = R
    
    x_app = np.arange(0, Rmax, Rmax/1000)
    plt.figure(figsize=(8, 8))
    plt.hist(R_data, np.arange(Rmax), density = True)
    plt.plot(x_app, R_loi(x_app, p, q), color='red', linewidth=2)
    
    #ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    #ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    plt.yticks(fontsize=12)
    plt.xlabel(r'R', fontsize=20)
    plt.ylabel(r'Frequance', fontsize=20)
    plt.grid(which='both')
    #plt.title(r'(%i, %i)' %(p, q), fontsize=20)
```

```python
R_graphique(10, 10)
R_graphique(20, 40)
```

### S6
Créer un programme en Python qui trace sur un même graphique 5 réalisations de (n, Sn(ω)) sur
l’intervalle [0, 1000].

```python
def hops(n):
    
    chemin = np.random.choice([-1, 1], n)

    return chemin
```

```python
def marches_aleatoires(hops, fig):
    
    S = np.zeros(hops.size+1)
    
    for i in range(hops.size):
            S[i+1] = S[i]+hops[i]
            
    plt.plot(np.arange(hops.size+1), S, markersize=6, color = 'r', linewidth = 2)
    #ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    #ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    plt.yticks(fontsize=12)
    plt.xlabel(r'n', fontsize=20)
    plt.ylabel(r'$S_n(\omega)$', fontsize=20)
    plt.grid(which='both')
```

```python
fig = plt.figure(figsize=(8, 8))
for i in range(5):
    marches_aleatoires(hops(1000), fig)
```

### S7

En simulant un grand nombre de marches, tracer la fonction de répartition empirique du premier
temps de retour en 0 de la sauterelle. On superposera sur le même graphique la fonction de répartition
théorique que l’on calculera numériquement à l’aide de la formule obtenue en T6.

```python
def pr_ret(N):
    ch = hops(N)
    S=0
    for i in range(N):
        S += ch[i]
        if S == 0:
            return i+1
    return -10

def pr_ret_modulation(N, n):
    stat = np.zeros(N, dtype = int)
    Smax = 0
    
    for i in range(N):
        stat[i] = pr_ret(n)
        if(stat[i] > Smax):
            Smax = stat[i]
        
    pr_ret_loi = np.array([1/(2*k-1) * math.comb(2*k-1, k) *1 / 2**(2*k-1)
                            for k in range(1, Smax//2)])
    plt.figure(figsize=(8, 8))
    plt.hist(stat, np.arange(Smax), density = True)
    plt.plot(np.arange(1, Smax//2), pr_ret_loi, color='red', linewidth=2)
    plt.yticks(fontsize=12)
    plt.xlabel(r'S', fontsize=20)
    plt.ylabel(r'Frequance', fontsize=20)
    plt.grid(which='both')
```

```python
pr_ret_modulation(500, 500)
```
