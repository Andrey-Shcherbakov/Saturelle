### S1. 
Écrire un programme qui prend en entrée un couple *(p, q)* avec *p ≤ q* et qui renvoie un tableau
numpy de taille *p+q* qui code un chemin de *(0, 0) à (p, q)* uniforme parmi tous les chemins possibles. On
pourra d’abord coder une fonction qui code (par convention) Nord=1 et Est=0 dans le tableau numpy.
Indication: on pourra utiliser la méthode *np.random.permutation* pour générer une permutation
aléatoire uniforme. 

```python
def chemin(p, q):
    if p > q:
        raise Exception("Sorry, we need p <= q")
    
    chemin = np.zeros(p+q, dtype=int)

    for i in range(q):
        chemin[i] = 1
    
    chemin = np.random.permutation(chemin)
        
    return chemin

p = (int)(input())
q = (int)(input())

print(chemin(p, q))
```

### S3.
Implémenter une fonction qui prend en entrée un couple *(p, q)* avec *p ≤ q* et qui renvoie un
chemin de *(0,0) à (p, q)* qui ne traverse pas la première bissectrice en utilisant la méthode du rejet
(qui consiste à générer des chemins avec le code *S1* et à conserver le premier qui ne traverse pas la
première bissectrice). Justifier rapidement que cette algorithme de rejet renvoie bien un chemin qui suit
la mesure uniforme sur les chemins qui ne traversent pas la première bissectrice.

```python
def verif(chemin): #la fonction qui vérifier si le chemin ne traverse pas la bissectrice 
    sum = chemin[0]
    for i in range(chemin.size-1):
        if(i > 0 and sum <= (i-i%2)/2): #si il traverse
            #print(i)
            return False
        sum = sum + chemin[i+1]
    return True

def rejet(p, q): #la fonction d'exercise
    if p > q:
        raise Exception("Sorry, we need p <= q")
    R = 0 #comte des rejets
    ch = chemin(p, q)
    while not verif(ch):
        ch = chemin(p, q)
        R = R+1
    
    #print(R)
    return ch
        
```

De point *T4* on à que toutes les chemins qui ne traverse pas la bissectrice sont uniformement (avec la même probabilité) distribués entres toutes les autres chemins (quand nous les choisissons) et notre algorithme de rejet suivi exectament le conditions comme dans l'exercise *T4*

### S5.
Notons *R* le nombre de rejets réalisés lors de la génération d’un chemin qui ne traverse pas la
première bissectrice selon l’algorithme implémenté en *S3*. D’après la question *T4*, quelle est la loi de
*R* ? Vérifier numériquement ce résultat en générant un grand nombre de tels chemin et en affichant
l’histogramme de la loi empirique de *R* que l’on superposera avec la loi théorique. On affichera les
résultats pour *(p = 10, q = 10)* puis pour *(p = 20, q = 40)*.

#### Solution
On a de point **T4** que la formule $$P(A) = \frac{q+1-p}{q+1}$$ de ça on peut facilement déduire que $$P(R) = (1-P(A))^{R-1}*P(A)$$ **Explication**: le R-ième iteration d'algorithme nous donne le **premiére** fois, quand il ne traverse pas la bissecrtice - c'est à dire que il ne traverse pas le premiére *R-1* fois et traverse le R-iéme 

```python
def R_loi(R, p, q): # la fonction qui returne la probabilité de R
    prob_A = (q+1-p)/(q+1)
    return (1 - prob_A)**(R-1) * prob_A


def R_graphique(p, q):
    if p > q:
        raise Exception("Sorry, we need p <= q")
    
    R_data = np.zeros(1000)
    Rmax = 0 # compte de R maximal pour facilite le construction de graphiqe
    
    for i in range(1000): # nombre 1000 ici était choisi pour les reasons du temps, de computation et pour avoir de resultat plus visible
        R = 0
        ch = chemin(p, q)
        while not verif(ch):
            ch = chemin(p, q)
            R = R+1
        if Rmax < R:
            Rmax = R
        R_data[i] = R
    
    x_app = np.arange(0, Rmax, Rmax/1000)
    plt.figure(figsize=(8, 8))
    plt.hist(R_data, np.arange(Rmax), density = True, label = 'la loi emperique')
    plt.plot(x_app, R_loi(x_app, p, q), color='red', linewidth=2, label = 'la loi théoretique')
    
    plt.xlabel(r'R', fontsize=18)
    plt.ylabel(r'P(R)', fontsize=18)
    plt.grid(which='both')
    plt.legend(loc = 'center right')
    plt.title(r'La distribution de R pour (%i, %i)' %(p, q), fontsize=20)
```

```python
R_graphique(10, 10)
R_graphique(20, 40)
```

## 2 Déplacement de la sauterelle sur une ligne

On peut reprendre les résultats de la partie précédente pour résoudre cette partie. On imagine une

### S6
Créer un programme en *Python* qui trace sur un même graphique 5 réalisations de *(n, Sn(ω))* sur
l’intervalle [0, 1000].

```python
def hops(n): # cette fonction crée un massif de commands: 
             # Droit = +1, gauche = -1
    
    chemin = np.random.choice([-1, 1], n)

    return chemin
```

```python
def marches_aleatoires(hops, fig): # elle designe le marches 
                                   #(execute des commands)
    
    S = np.zeros(hops.size+1)
    
    for i in range(hops.size):
            S[i+1] = S[i]+hops[i]
            
    plt.plot(np.arange(hops.size+1), S, markersize=6, linewidth = 1)
    plt.xlabel(r'n', fontsize=18)
    plt.ylabel(r'$S_n(\omega)$', fontsize=18)
    plt.grid(which='both')
    plt.title(r'Les marches aléatoires pour n = %i' %hops.size, fontsize=20)
```

```python
fig = plt.figure(figsize=(8, 8))
for i in range(5):
    marches_aleatoires(hops(1000), fig)
```

### S7

En simulant un grand nombre de marches, tracer la fonction de répartition empirique du premier
temps de retour en 0 de la sauterelle. On superposera sur le même graphique la fonction de répartition
théorique que l’on calculera numériquement à l’aide de la formule obtenue en *T6*.

**Solution**

De point *T6* on a obtenue la formule pour 2n: $$ \mathbb{P}_{2n} = \frac{1}{2^{2n-1}}\frac{1}{2n-1}\tbinom{2n-1}{n}$$

```python
def pr_ret(N): #une fonction qui renvoi le numero de premiére retour ou -10, si il n'est pas retourné pendant la simulation
    ch = hops(N)
    S=0
    for i in range(N):
        S += ch[i]
        if S == 0:
            return i+1
    return -10

def pr_ret_modulation(N, n): #modulation de exercise
    stat = np.zeros(N, dtype = int)
    Smax = 0
    
    for i in range(N):
        stat[i] = pr_ret(n)
        if(stat[i] > Smax):
            Smax = stat[i]
        
    pr_ret_loi = np.array([1/(2*k-1) * 1 / 2**(2*k-1) * math.comb(2*k-1, k)
                            for k in range(1, Smax)]) # le loi d'exercise T11
    plt.figure(figsize=(8, 8))
    plt.hist(stat, np.arange(Smax), density = True, label = 'la loi emperique')
    plt.plot(np.arange(1, Smax), pr_ret_loi, color='red', 
             linewidth=2, label = 'la loi théoretique')
    
    plt.xlabel(r'n', fontsize=18)
    plt.ylabel(r'Probabilité de revenir', fontsize=18)
    plt.grid(which='both')
    plt.legend(loc = 'center right')
```

```python
pr_ret_modulation(1000, 400)
```

### S8

Créer un programme en Python qui trace sur un même graphique 4 réalisations *(n, Sn(ω))* sur
l’intervalle [0, 1000]. On pourra utiliser la librairie mpl_toolkits1 pour réaliser les graphiques 3D.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
def marches_aleatoires3D(n, ax):
    plt.figure(1)
    
    u = hops(n) # on utilise le question T12: u et v sont independantes et avec le distribution uniforme: P(U =1)=P(U=-1)=1/2 pour V - pareil
    v = hops(n)
    
    x = (u+v)/2
    y = (u-v)/2
    
    S_x = np.zeros(n+1) 
    S_y = np.zeros(n+1)
    
    for i in range(n):
        S_x[i+1] = S_x[i]+x[i]
        S_y[i+1] = S_y[i]+y[i]
        
    N = np.arange(n+1)
    ax.plot3D(N, S_x, S_y)
    ax.set_xlabel('n', fontsize=18)
    ax.set_ylabel('$S_n(X_n)$', fontsize=18)
    ax.set_zlabel('$S_n(Y_n)$', fontsize=18)
    
    plt.grid()
    
    plt.figure(2) # seulement sur axis x
    plt.plot(N, S_x)
    plt.xlabel(r'n', fontsize=18)
    plt.ylabel(r'$S_n(X_n)$', fontsize=18)
    plt.grid(which='both')
    
    plt.figure(3)
    plt.plot(N, S_y) #seulement sur axis y
    plt.xlabel(r'n', fontsize=18)
    plt.ylabel(r'$S_n(Y_n)$', fontsize=18)
    plt.grid(which='both')
```

```python
fig = plt.figure(figsize=(8, 8)) #3d visualisation
ax  = fig.add_subplot(111, projection='3d')
fig1 = plt.figure(figsize=(8, 8)) # profection sur axis x
fig2 = plt.figure(figsize=(8, 8)) # profection sur axis y
for i in range(4):
    marches_aleatoires3D(1000, ax)
```
